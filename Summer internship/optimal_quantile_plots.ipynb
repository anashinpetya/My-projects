```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def four_pl(a,b,c,d,q):
    return a+(d-a)/(1+(q/c)**b)
```

```python
df=pd.read_excel("quantile_values.xlsx")
```

```python
df.set_index("Quantile")
```

```python
function_values_am=[]
for q in np.linspace(0,1,1000):
    function_values_am.append(four_pl(81896.5008019733,3.5814868213337,0.394216025142771,141402.51881059,q))
```

```python
function_values_pm=[]
for q in np.linspace(0,1,1000):
    function_values_pm.append(four_pl(79750.8318471373,3.27095965837775,0.404451597629991,131926.651008284,q))
```

```python
df_function=pd.DataFrame(data=[function_values_am,function_values_pm]).T
df_function.columns=["am","pm"]
df_function["Quantile"]=np.linspace(0,1,1000)
df_function=df_function.set_index("Quantile")
```

```python
sns.set(style='white',context='paper',rc={'axes.edgecolor': '#F2F2F2','axes.facecolor':'#F2F2F2', 'figure.facecolor':'#F2F2F2'})
plt.figure(figsize=(8, 7))
sns.lineplot(data=df_function["am"],label="Model",zorder=1)
sns.scatterplot(data=df.set_index("Quantile")["am"],color="black",label="Iteration's data",zorder=2,s=90)
plt.scatter(x=0.32,y=four_pl(81896.5008019733,3.5814868213337,0.394216025142771,141402.51881059,0.32), color="orange", label="Quantile = 0.32",zorder=3,s=200)
plt.xlabel("Quantile",fontsize=16)
plt.ylabel("Number of car trips",fontsize=16)
plt.title("Relationship between quantile and number of trips (am data)",fontsize=18,pad=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=14)
plt.grid(color="#BABABA",linewidth=1)
sns.despine()
plt.savefig(r"C:\Users\anash\OneDrive - Tartu Ülikool\Aalto internship\Poster\Plots\quantile_curve_am.svg",bbox_inches='tight')
```

```python
sns.set(style='white',context='paper',rc={'axes.edgecolor': '#F2F2F2','axes.facecolor':'#F2F2F2', 'figure.facecolor':'#F2F2F2'})
plt.figure(figsize=(8, 7))
sns.lineplot(data=df_function["pm"],label="Model",zorder=1)
sns.scatterplot(data=df.set_index("Quantile")["pm"],color="black",label="Iteration's data",zorder=2,s=90)
plt.scatter(x=0.32,y=four_pl(79750.8318471373,3.27095965837775,0.404451597629991,131926.651008284,0.32), color="orange", label="Quantile = 0.32",zorder=3,s=200)
plt.xlabel("Quantile",fontsize=16)
plt.ylabel("Number of car trips",fontsize=16)
plt.title("Relationship between quantile and number of trips (pm data)",fontsize=18,pad=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=14)
plt.grid(color="#BABABA",linewidth=1)
sns.despine()
plt.savefig(r"C:\Users\anash\OneDrive - Tartu Ülikool\Aalto internship\Poster\Plots\quantile_curve_pm.svg",bbox_inches='tight')
```

```python
sns.set(style='white',context='paper',rc={'axes.edgecolor': '#F2F2F2','axes.facecolor':'#F2F2F2', 'figure.facecolor':'#F2F2F2'})
plt.figure(figsize=(8, 7))
sns.lineplot(data=df_function["am"],label="Model",zorder=1)
sns.scatterplot(data=df.set_index("Quantile")["am"],color="black",label="Iteration's data",zorder=2,s=90)
plt.scatter(x=0.377169992,y=four_pl(81896.5008019733,3.5814868213337,0.394216025142771,141402.51881059,0.377169992), color="orange", label="Optimal quantile",zorder=3,s=200)
plt.xlabel("Quantile",fontsize=16)
plt.ylabel("Number of car trips",fontsize=16)
plt.title("Identification of optimal quantile, am data",fontsize=18,pad=15)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=14)
plt.grid(color="#BABABA",linewidth=1)
sns.despine()
plt.savefig("optimal_quantile_curve_am.svg")
```
