```python
df=df.sort_values(by=["time_system_ts","id"]).reset_index().drop(columns=["index"])
```

```python
#time difference in seconds between start and end points
df["time_diff"]=(df["time_system_ts"].shift(-1)-df["time_system_ts"])/np.timedelta64(1,'s')
```

```python
#if location has not changed, then speed is equal to 0
df["end_x"]=df["x"].shift(-1)
df["end_y"]=df["y"].shift(-1)
```

```python
index_unchanged_loc=df[df["x"]==df["end_x"]][df[df["x"]==df["end_x"]]["y"]==df[df["x"]==df["end_x"]]["end_y"]].index
```

```python
#number of filling gaps
df.loc[index_unchanged_loc]["speed"].isna().sum()
```

```python
#fill the gaps where location has not changed
df.loc[df.loc[index_unchanged_loc]["speed"][df.loc[index_unchanged_loc]["speed"].isna()].index,"speed"]=0
```

```python
#time difference in seconds between previous and current points
df["prev_time_diff"]=(df["time_system_ts"]-df["time_system_ts"].shift(1))/np.timedelta64(1,'s')
```

```python
filled_distances["speed"]=filled_distances["speed_calc"]
filled_distances
```

```python
filled_distances=filled_distances.drop(columns="speed_calc")
filled_distances
```

```python
#drop old rows with nan values in speed
df=df.drop(index=df[df["id"].isin(filled_distances.id)].index)
df
```

```python
filled_distances["time_system_ts"]=pd.to_datetime(filled_distances["time_system_ts"])
```

```python
df_corr=pd.concat([df,filled_distances]).sort_values(by=["time_system_ts","id"])
```

```python
df_corr=df_corr.drop(columns=["fid","distance"])
```

```python
df_corr["time_diff"]=(df_corr["time_system_ts"].shift(-1)-df_corr["time_system_ts"])/np.timedelta64(1,'s')
df_corr["prev_time_diff"]=(df_corr["time_system_ts"]-df_corr["time_system_ts"].shift(1))/np.timedelta64(1,'s')
```

```python
#it makes sense to remove these rows because they are duplicates (time_diff=0)
df_corr=df_corr.drop(index=df_corr[df_corr["speed"]==np.inf].index)
df_corr
```
